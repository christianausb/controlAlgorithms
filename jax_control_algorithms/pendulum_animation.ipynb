```python
import matplotlib.pyplot as plt
import numpy as np

def plot_cart_pendulum(cart_position, pendulum_angle):
    # Set up the figure and axis
    fig, ax = plt.subplots()

    # Define the cart dimensions
    cart_width = 0.4
    cart_height = 0.2

    # Draw the cart
    cart = plt.Rectangle((cart_position - cart_width / 2, 0), cart_width, cart_height, fill=True, color='blue')
    ax.add_patch(cart)

    # Define the pendulum dimensions
    pendulum_length = 1.0  # Length of the pendulum rod

    # Calculate the pendulum bob position
    pendulum_bob_x = cart_position + pendulum_length * np.sin(pendulum_angle)
    pendulum_bob_y = cart_height + pendulum_length * np.cos(pendulum_angle)

    # Draw the pendulum rod
    pendulum_rod, = ax.plot([cart_position, pendulum_bob_x], [cart_height, pendulum_bob_y], color='black', linewidth=2)

    # Draw the pendulum bob
    pendulum_bob = plt.Circle((pendulum_bob_x, pendulum_bob_y), 0.1, fill=True, color='red')
    ax.add_patch(pendulum_bob)

    # Set axis limits and labels
    ax.set_xlim(-2, 2)
    ax.set_ylim(0, 2)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('Position')
    ax.set_ylabel('Height')

    # Display the plot
    plt.show()

# Example usage
cart_position = 0.0  # Specify the cart position
pendulum_angle = np.pi / 4  # Specify the pendulum angle in radians
plot_cart_pendulum(cart_position, pendulum_angle)
```

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np

# Function to update the animation frame
def update(frame):
    ax.clear()
    
    # Define the cart dimensions
    cart_width = 0.4
    cart_height = 0.2

    # Draw the cart
    cart = plt.Rectangle((cart_position[frame] - cart_width / 2, 0), cart_width, cart_height, fill=True, color='blue')
    ax.add_patch(cart)

    # Define the pendulum dimensions
    pendulum_length = 1.0  # Length of the pendulum rod

    # Calculate the pendulum bob position
    pendulum_bob_x = cart_position[frame] + pendulum_length * np.sin(pendulum_angle[frame])
    pendulum_bob_y = cart_height + pendulum_length * np.cos(pendulum_angle[frame])

    # Draw the pendulum rod
    pendulum_rod, = ax.plot([cart_position[frame], pendulum_bob_x], [cart_height, pendulum_bob_y], color='black', linewidth=2)

    # Draw the pendulum bob
    pendulum_bob = plt.Circle((pendulum_bob_x, pendulum_bob_y), 0.1, fill=True, color='red')
    ax.add_patch(pendulum_bob)

    # Set axis limits and labels
    ax.set_xlim(-2, 2)
    ax.set_ylim(0, 2)
    ax.set_aspect('equal', adjustable='box')
    ax.set_xlabel('Position')
    ax.set_ylabel('Height')

# Example time-series data for cart position and pendulum angle (replace with your data)
cart_position = np.linspace(-1, 1, 100)
pendulum_angle = np.sin(np.linspace(0, 2 * np.pi, 100))

# Create the figure and axis
fig, ax = plt.subplots()

# Create the animation
ani = FuncAnimation(fig, update, frames=len(cart_position), repeat=False, interval=100)

# Save the animation as an MP4 file
ani.save('cart_pendulum_animation.mp4', writer='ffmpeg')

# Display the saved animation
from IPython.display import Video
Video("cart_pendulum_animation.mp4")
```
